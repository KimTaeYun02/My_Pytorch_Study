## DataLoader
PytorchÎäî DataLoaderÏù¥ÎùºÍ≥† ÌïòÎäî utilityÎ•º Ï†úÍ≥µÌïúÎã§.


Í∞ÑÎã®ÌïòÍ≤å ÎßêÌïòÎ©¥ DataLoader Í∞ùÏ≤¥Îäî ÌïôÏäµÏóê Ïì∞Ïùº data Ï†ÑÏ≤¥Î•º Î≥¥Í¥ÄÌñàÎã§Í∞Ä, train Ìï®ÏàòÍ∞Ä batchÎ•º ÏöîÍµ¨ÌïòÎ©¥ batch size Í∞úÏàòÎßåÌÅº Îç∞Ïù¥ÌÑ∞Î•º Í∫ºÎÇ¥ÏÑú Ï§ÄÎã§.

    * Ïù¥ ÏΩîÎìúÏóêÏÑúÎäî Ïù¥ÎØ∏ÏßÄÏùò Í≤ΩÎ°úÎ•º ÏãùÎ≥ÑÌïòÍ≥†, read_imageÎ•º ÏÇ¨Ïö©ÌïòÏó¨ Ïù¥ÎØ∏ÏßÄÎ•º ÌÖêÏÑúÎ°ú Î≥ÄÌôòÌïòÍ≥†, self.img_labelsÏùò csv Îç∞Ïù¥ÌÑ∞Î°ú Î∂ÄÌÑ∞ Ìï¥ÎãπÌïòÎäî Ï†ïÎãµ(label)ÏùÑ Í∞ÄÏ†∏Ïò§Í≥†, Î≥ÄÌòï(transform)Ìï®ÏàòÎì§ÏùÑ Ìò∏Ï∂úÌïú Îí§, ÌÖêÏÑú Ïù¥ÎØ∏ÏßÄÏôÄ labelÏùÑ dicÌòïÏúºÎ°ú Î∞òÌôò
    * tensorÎ°ú Î≥ÄÌôòÌï¥Ï£ºÏñ¥Ïïº Ìï®.
  
#### image = read_image(img_path)

- jpeg or png Ïù¥ÎØ∏ÏßÄÎ•º tensor(0~255)Î°ú Î∂àÎü¨Ïò¥.

#### label = self.img_labels.iloc[idx, 1]
- __ init __ Î∂ÄÎ∂ÑÏóêÏÑú pandasÎ°ú csvÌååÏùºÏùÑ Í∞ÄÏ†∏Ïò§Í≥† img_labelsÏóê Ï†ÄÏû•Ìï®.
- Ïù¥Îïå 0Î≤àÏß∏ columnÏù¥ file_name, 1Î≤àÏß∏ columnÏù¥ labelÏûÑ
- __.iloc[row, column]__Ïù¥ÎØÄÎ°ú idxÌñâÏùò labelÏùÑ label Î≥ÄÏàòÏóê Ï†ÄÏû•


```python
import os
import pandas as pd
from torchvision.io import read_image
from torch.utils.data import DataLoader
```

```python
class CustomImageDataset(Dataset):
    
    def __init__(self, annotations_file, img_dir,
                 transform=None, target_transform=None):
        self.img_labels = pd.read_csv(annotations_file,
                                      names=['file_name', 'label'])
        self.img_dir = img_dir
        self.transform = transform
        self.target_transform = target_transform
        
    #datasetÏùò Ï†ÑÏ≤¥ Í∞úÏàòÎ•º ÏïåÎ†§Ï§ÄÎã§
    def __len__(self):
        return len(self.img_labels)
    
    #parameter
    def __getitem__(self, idx):
        # Ïù¥ÎØ∏ÏßÄ Í≤ΩÎ°ú
        img_path = os.path.join(self.img_dir, self.img_labels.iloc[idx, 0])
        
        # imageÎ•º tensorÎ°ú load
        image = read_image(img_path)
        # label Ï†ÄÏû•
        # idxÎ≤àÏß∏ row Ï§ë 1Î≤àÏß∏ col (idx,1)Ïùò Í∞íÏùÑ labelÏóê Ï†ÄÏû• 
        label = self.img_labels.iloc[idx, 1]
        
        # imageÏôÄ label transform 
        if self.transform:
            image = self.transform(image)
        if self.target_transform:
            label = self.target_transform(label)
        
        # image, labelÏùÑ return
        return image, label
```

* __ init __
 **Î™®Îç∏ÏóêÏÑú ÏÇ¨Ïö©Ìï† module, layerÎ•º Ï†ïÏùòÌïúÎã§.** moduleÎßå Ï†ïÏùòÌï† Ïàò ÎèÑ, activation function Îì±ÏùÑ Ï†ÑÎ∂Ä Ï†ïÏùòÌï† Ïàò ÏûàÎã§.

* forward()
 Î™®Îç∏ÏóêÏÑú ÏàòÌñâÌï¥ÏïºÌïòÎäî Í≥ÑÏÇ∞ÏùÑ Ï†ïÏùòÌïúÎã§(Î≥¥ÌÜµ trainÌï† Îïå). forward Ïó∞ÏÇ∞Ïù¥ÎûÄ H(x)ÏãùÏóê input data xÎ°ú Î∂ÄÌÑ∞ ÏòàÏ∏°Îêú yÎ•º ÏñªÎäî Í≤ÉÏù¥Îã§.

```python
import os
import torch
from torch import nn
from torch.utils.data import DataLoader
from torchvision import datasets, transforms
```

```python
# cudaÏù∏ÏßÄ cpuÏù∏ÏßÄ ÌôïÏù∏
device = "cuda" if torch.cuda.is_available() else "cpu"
print(f"Using {device} device")
```

* `nn.Flatten()`
28 * 28Ïùò 2D Ïù¥ÎØ∏ÏßÄÎ•º 784 ÌîΩÏÖÄ Í∞íÏùÑ Í∞ñÎäî Î∞∞Ïó¥Î°ú Î≥ÄÌôò. ÌèâÌèâÌïòÍ≤å Ìé¥Ï§ÄÎã§Îäî ÎäêÎÇå.

* `nn.Linear(in_feature, out_feature)`
ÏÑ†Ìòï Í≥ÑÏ∏µÏùÄ Ï†ÄÏû•Îêú Í∞ÄÏ§ëÏπò(w)ÏôÄ Ìé∏Ìñ•(b)Î•º ÏÇ¨Ïö©ÌïòÏó¨ ÏûÖÎ†•Ïóê ÏÑ†Ìòï Î≥ÄÌôòÏùÑ Ï†ÅÏö©ÌïòÎäî Î™®Îìà. ÏûÖÎ†• Îç∞Ïù¥ÌÑ∞Í∞Ä 28 * 28 Ï¶â, 784Ïù¥ÎØÄÎ°ú in_featureÏù¥ 784. Îëê Î≤àÏß∏ Linear layerÏùÄ input featureÍ∞Ä 512Ïù¥ÎØÄÎ°ú 512Ïù∏ Í≤É. ÎßàÏßÄÎßâ out_featureÏùÄ Î∂ÑÎ•òÌï¥ÏïºÌïòÎäî classÍ∞Ä 10Í∞úÏù¥ÎØÄÎ°ú 10Ïù¥Îã§.

* `nn.ReLU()`
ÌôúÏÑ±Ìôî Ìï®ÏàòÎäî ÏûÖÎ†•Í≥º Ï∂úÎ†• ÏÇ¨Ïù¥Ïóê Îì§Ïñ¥Í∞ê.

* `nn.Sequntial`
ÏàúÏÑúÎ•º Í∞ñÎäî moduleÏùò Ïª®ÌÖåÏù¥ÎÑà. ÏïÑÎûòÏùò linear_relu_stackÏ≤òÎüº Ïã†Í≤ΩÎßùÏùÑ Îπ†Î•¥Í≤å ÎßåÎì§ Ïàò ÏûàÎã§. piplineÍ∞ôÏùÄ ÎäêÎÇå

```python
class NeuralNetwork(nn.Module):
    def __init__(self):
        #Î∂ÄÎ™® ÌÅ¥ÎûòÏä§ Ìò∏Ï∂ú(init)
        super(NeuralNetwork, self).__init__()
        
        self.flatten = nn.Flatten()
        self.linear_relu_stack = nn.Sequential(
            nn.Linear(28*28, 512),
            nn.ReLU(),
            nn.Linear(512, 512),
            nn.ReLU(),
            nn.Linear(512, 10),
        )

    def forward(self, x):
        x = self.flatten(x)
        logits = self.linear_relu_stack(x)
        return logits
```

```python
# Î™®Îç∏Ïùò Íµ¨Ï°∞ ÌôïÏù∏ 
model = NeuralNetwork().to(device)
print(model)
```

## Î™®Îç∏ Îß§Í∞úÎ≥ÄÏàò ÏµúÏ†ÅÌôî

Î™®Îç∏(NeuralNetwork)Í≥º Îç∞Ïù¥ÌÑ∞(DataLoader)Í∞Ä Ï§ÄÎπÑÎêòÏóàÏúºÎãà Îç∞Ïù¥ÌÑ∞Ïóê Îß§Í∞úÎ≥ÄÏàòÎ•º ÏµúÏ†ÅÌôîÌïòÏó¨ Î™®Îç∏ÏùÑ ÌïôÏäµÌï† Ï∞®Î°Ä. Î™®Îç∏ÏùÑ ÌïôÏäµÌïòÎäî Í≥ºÏ†ïÏùÄ Î∞òÎ≥µÏ†ÅÏù∏ Í≥ºÏ†ïÏùÑ Í±∞ÏπúÎã§. 


Í∞Å Îã®Í≥Ñ(epoch)ÏóêÏÑúÎäî 1)Î™®Îç∏Ïùò targetÍ∞íÏùÑ ÏòàÏ∏°ÌïòÍ≥†, 2)y_perdictÍ∞íÍ≥º y_targetÏÇ¨Ïù¥Ïùò lossÎ•º Í≥ÑÏÇ∞ÌïòÍ≥† 3)Í≤ΩÏÇ¨ÌïòÍ∞ïÎ≤ïÏùÑ ÌÜµÌï¥ ÌååÎùºÎØ∏ÌÑ∞Îì§ÏùÑ optimizeÌïúÎã§.


Î™®Îç∏ÏùÑ ÌïôÏäµÌïòÎ†§Î©¥ ÏÜêÏã§ Ìï®Ïàò(Loss function)Í≥º ÏòµÌã∞ÎßàÏù¥Ï†Ä(Optimizer)Í∞Ä ÌïÑÏöîÌïòÎã§.

---

#### loss funtunctionÍ≥º optimizer
 Îî•Îü¨ÎãùÏóêÏÑúÎäî ÌïôÏäµÏù¥ ÏµúÎåÄÌïú ÌãÄÎ¶¨ÏßÄ ÏïäÎäî Î∞©Ìñ•ÏúºÎ°ú ÌïôÏäµÌï¥ ÎÇòÍ∞ÄÏïº ÌïòÎäîÎç∞. Ïó¨Í∏∞ÏÑú ÏñºÎßàÎÇò ÌãÄÎ¶¨ÎäîÏßÄ **loss(Ïò§Ï∞®)** Î•º Í≥ÑÏÇ∞ ÌïòÎäî Ìï®ÏàòÍ∞Ä **lossfuntion(ÏÜêÏã§Ìï®Ïàò)** Ïù¥Îã§. Ïù¥ **loss funtion** Ïùò ÏµúÏÜåÍ∞íÏùÑ Ï∞æÎäî Í≤ÉÏùÑ ÌïôÏäµÏùò Î™©ÌëúÎ°ú ÌïòÎäîÎç∞ Ïó¨Í∏∞ÏÑú ÏµúÏÜåÍ∞íÏùÑ Ï∞æÏïÑÍ∞ÄÎäî Í≤ÉÏù¥ **Optimization(ÏµúÏ†ÅÌôî)** Ïù¥ÎùºÍ≥† ÌïòÍ≥† Ïù¥Î•º ÏàòÌñâÌïòÎäî ÏïåÍ≥†Î¶¨Ï¶òÏù¥ **Optimizer(ÏµúÏ†ÅÌôî ÏïåÍ≥†Î¶¨Ï¶ò)** Ïù¥Îã§.

```python
def train_loop(dataloader, model, loss_fn, optimizer):
    size = len(dataloader.dataset)
    for batch, (X, y) in enumerate(dataloader):
        # ÏòàÏ∏°(prediction)Í∞í Í≥ÑÏÇ∞
        pred = model(X)
        loss = loss_fn(pred, y)

        # Ïó≠Ï†ÑÌåå
        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        if batch % 100 == 0:
            loss, current = loss.item(), batch * len(X)
            print(f"loss: {loss:>7f}  [{current:>5d}/{size:>5d}]")
```

```python
def test_loop(dataloader, model, loss_fn):
    size = len(dataloader.dataset)
    num_batches = len(dataloader)
    test_loss, correct = 0, 0

    with torch.no_grad():
        for X, y in dataloader:
            pred = model(X)
            test_loss += loss_fn(pred, y).item()
            correct += (pred.argmax(1) == y).type(torch.float).sum().item()

    test_loss /= num_batches
    correct /= size
    print(f"Test Error: \n Accuracy: {(100*correct):>0.1f}%, Avg loss: {test_loss:>8f} \n")
```
